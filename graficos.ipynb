```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
from tp1_greedy import obtener_suma_ponderada
from main import obtener_datos
```

```python
data_sizes = list(range(0, 100000, 1000))
puntos = []

for n in data_sizes:
    # Vamos a correr varias veces el algoritmo para sacar un timepo promedio
    tiempos_aux = [] 
    for i in range(5): 
        # Creamos una lista random de valores de tiempo y peso de las batallas
        datos_batallas = []
        for j in range(n):
            datos_batallas.append((random.randint(1, 1000),random.randint(1, 1000)))
        
        inicio_tiempo = time.time()  # Iniciar medicin de tiempo

        suma_ponderada = obtener_suma_ponderada(datos_batallas)

        fin_tiempo = time.time()  # Finalizar medicin de tiempo
        tiempo_total = fin_tiempo - inicio_tiempo  # Calcular tiempo total
        tiempos_aux.append(tiempo_total)

    puntos.append(statistics.mean(tiempos_aux))

#Graficamos los puntos
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(data_sizes, puntos, marker='o')
ax.grid()
ax.set_xlabel('Tamao de los datos')
ax.set_xticks([1000,5000,10000,20000,50000,100000])
ax.set_ylabel('Tiempo de ejecucin')
ax.set_title('Tiempo de ejecucin segn el tamao de los datos')
ax.set_facecolor('#E7E6E6')
ax.grid(color='white', alpha=0.8)

# Generar puntos de la complejidad O(n log n)
# complejidad_n_log_n = [n * np.log(n) for n in data_sizes]
# plt.plot(data_sizes, complejidad_n_log_n, linestyle='--', color='r', label='O(n log n)')
```
