### Intervalo de 1000

```python
data_sizes = list(range(1, 100000, 1000))
puntos = []

for n in data_sizes:
    # Vamos a correr varias veces el algoritmo para sacar un timepo promedio
    tiempos_aux = [] 
    for i in range(5): 
        # Creamos una lista random de valores de tiempo y peso de las batallas
        datos_batallas = []
        for j in range(n):
            datos_batallas.append((random.randint(1, 1000),random.randint(1, 1000)))
        
        inicio_tiempo = time.time()  # Iniciar medicin de tiempo

        suma_ponderada = obtener_suma_ponderada(datos_batallas)

        fin_tiempo = time.time()  # Finalizar medicin de tiempo
        tiempo_total = fin_tiempo - inicio_tiempo  # Calcular tiempo total
        tiempos_aux.append(tiempo_total)

    puntos.append(statistics.mean(tiempos_aux))

#Graficamos los puntos
fig, ax = plt.subplots(1, 2, figsize=(15, 6))


complejidad_n_log_n = data_sizes * np.log(data_sizes)
ax[0].set_title("n * log(n)")
ax[0].plot(data_sizes, complejidad_n_log_n, linestyle='--', color='#B31312')
ax[0].set_xlabel('Tamao de los datos')
ax[0].set_ylabel('Tiempo de ejecucin [s]')
ax[0].set_facecolor('#E7E6E6')
ax[0].grid(color='white', alpha=0.8)

ax[1].plot(data_sizes, puntos, marker='o', color='#B31312')
ax[1].grid()
ax[1].set_xlabel('Tamao de los datos')
ax[1].set_ylabel('Tiempo de ejecucin [s]')
ax[1].set_title('Tiempo de ejecucin segn el tamao de los datos')
ax[1].set_facecolor('#E7E6E6')
ax[1].grid(color='white', alpha=0.8)
```

### Medicin con todos los criterios

```python
data_sizes = list(range(1, 1000, 10))

puntos_ratio = []
puntos_peso = []
puntos_tiempo = []

for n in data_sizes:
    # Creamos una lista random de valores de tiempo y peso de las batallas
    datos_batallas = []
    for j in range(n):
        datos_batallas.append((random.randint(1, 1000), random.randint(1, 1000)))

    # Obtener suma ponderada por ratio
    suma_ponderada = obtener_suma_ponderada(datos_batallas)
    puntos_ratio.append(suma_ponderada)
    # Obtener suma ponderada por peso
    suma_ponderada_por_peso = obtener_suma_ponderada_por_peso(datos_batallas)
    puntos_peso.append(suma_ponderada_por_peso)
    # Obtener suma ponderada por tiempo
    suma_ponderada_por_tiempo = obtener_suma_ponderada_por_tiempo(datos_batallas)
    puntos_tiempo.append(suma_ponderada_por_tiempo)

# Graficar los puntos
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(data_sizes, puntos_ratio, label='Ratio (peso/tiempo)', color='b')
ax.plot(data_sizes, puntos_peso, label='Peso', color='r')
ax.plot(data_sizes, puntos_tiempo, label='Tiempo', color='g')

ax.grid()
ax.set_xlabel('Tamao de los datos')
ax.set_ylabel('Suma ponderada')
ax.set_title('Comparacin del desempeo segn diferentes criterios')
ax.set_facecolor('#E7E6E6')
ax.grid(color='white', alpha=0.8)
ax.legend()
#ax.set_yticks(list(range(0, 1000000, 1000)))

plt.show()
```
